```python
etihad_airways = pd.read_csv('etihad_airways.csv')
qatar_airways = pd.read_csv('qatar_airways.csv')
emirates = pd.read_csv('emirates.csv')
```

```python
features = ['seat_comfort', 'cabin_service', 'food_bev', 'entertainment', 'ground_service', 'wifi', 'value_for_money']#, 'recommended', 'overall']


import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrames (replace these with your actual DataFrames)
# Assuming you have two DataFrames: qatar_airways and etihad_airways
# with columns 'seat_comfort', 'cabin_service', 'food_bev', 'entertainment', 'ground_service', 'wifi'
# and integer ratings from 1 to 5

# Calculate the percentage of each rating for each DataFrame
qatar_airways_percentages = qatar_airways[features].apply(lambda x: x.value_counts(normalize=True)) #* 100)
etihad_airways_percentages = etihad_airways[features].apply(lambda x: x.value_counts(normalize=True))# * 100)
emirates_airways_percentages = emirates[features].apply(lambda x: x.value_counts(normalize=True))# * 100)

# Plot subplots for each feature
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 8))

for i, feature in enumerate(features):
    row, col = divmod(i, 3)
    
    # Plot for Qatar Airways
    qatar_airways_percentages[feature].sort_index().plot(kind='bar', ax=axes[row, col], color='blue', alpha=0.5, label='Qatar Airways')
    
    # Plot for Etihad Airways
    etihad_airways_percentages[feature].sort_index().plot(kind='bar', ax=axes[row, col], color='orange', alpha=0.5, label='Etihad Airways')
    
    # Plot for Emirates 
    # emirates_airways_percentages[feature].sort_index().plot(kind='bar', ax=axes[row,col], color='red', alpha=0.3, label='Emirates')
    axes[row, col].set_title(f'{feature}')
    axes[row, col].set_xlabel('Rating')
    axes[row, col].set_ylabel('Percentage')
    axes[row, col].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
